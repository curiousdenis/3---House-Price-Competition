Here we use a <a href = https://towardsdatascience.com/methods-for-normality-test-with-application-in-python-bb91b49ed0f5 title = shapirto wilk test> shapiro wilk test</a> to check the normality of the data (quantiative). 

Below we can see again **all variables are not normally distributed** as the null hypothesis is rejected

```python
normal = pd.DataFrame(train[quantitative])
test_normality = lambda x: stats.shapiro(x.fillna(0))
normal = normal.apply(test_normality)
#to check the p-value of shapiro wilk
normal = normal.iloc[1]

#asssumption that it should be less than 0.05%
print((normal < 0.05).all())
```

```python
# A small example of anova_test
samples = []
sample_1 = train[train['Neighborhood'] == 'CollgCr']['SalePrice'].values
samples.append(sample_1)
sample_2 = train[train['Neighborhood'] == 'Veenker']['SalePrice'].values
samples.append(sample_2)
stats.f_oneway(*samples)[1] < 0.05
```

A one way **ANOVA** is used to compare two means (or more) from two (or more) independent (unrelated) groups using the F-distribution. The null hypothesis for the test is that the two means are equal. Therefore, a significant result means that the two means are unequal.

If one-way ANOVA reports a P value of <0.05, you reject the null hypothesis that all the data are sampled from populations with the same mean.

```python
def anova_test(data):
    df = pd.DataFrame()
    df['features'] = qualitative
    p_values = []
    for x in qualitative:
        samples = []
        for y in data[x].unique():
            sample = data[data[x] == y]['SalePrice'].values
            samples.append(sample)
        p_value = stats.f_oneway(*samples)[1]
        p_values.append(p_value)
    df['p_values'] = p_values
    return df.sort_values('p_values', ascending = False)
data = anova_test(train)
print(data[data['p_values'] > 0.05])
print('-----------------------------')
print(data)
```
